```python
df = pd.read_excel('実験集計表共有用.xlsx', sheet_name = "基本情報")
ori_df = df
df.columns
```

```python
df3 = ori_df[["調査ID", '性別', '年齢', '職業', '年収(総支給額)']]
df3['性別'] = df3['性別'].replace({'1. 男性': 1, '2. 女性': 0})
df3 = df3.rename(columns={'性別': '男性ダミー'})
df3['年齢'] = df3['年齢'].str[3:5]
df3['年齢'] = df3['年齢'].astype('int')
df3['年齢の2乗'] = pow(df3['年齢'], 2)
df3['年収(カテゴリ）'] = df3['年収(総支給額)'].str[:1]
df3['職業'] = df3['職業'].replace({'アルバイト/パート': '非正規雇用ダミー', '会社員': '正社員ダミー', '学生': '学生ダミー', '専業主婦・主夫': '主婦（主夫）ダミー', '無職' : '無職ダミー', '自営業/個人事業' : '自由業ダミー', '自由業' : '自由業ダミー'})
df3 = pd.get_dummies(df3, columns=['職業'],prefix='', prefix_sep='')
df3.head()
#df3.to_csv('data/preprocessed/df_attribute_dummy.csv')
```
