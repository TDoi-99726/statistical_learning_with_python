# データ読み込み

```python
#basicinfo = pd.read_csv("basicinfo.csv")
#basicinfo_ids = list(basicinfo["調査ID"])
#print(basicinfo_ids)

pcdf = pd.read_csv("/content/drive/Shareddrives/課金調査/コード/入力/PC.csv")
pcids = list(pcdf["調査ID"])
print(pcids)

phonedf = pd.read_csv("/content/drive/Shareddrives/課金調査/コード/入力/smartphone.csv")
phoneids = list(phonedf["調査ID"])
print(phoneids)
```

```python
phonedf
```

```python
pcdf
```

```python
#転置
pcdf = pcdf.drop(["タイムスタンプ","Unnamed: 0"],axis=1).set_index("調査ID")
phonedf = phonedf.drop(["タイムスタンプ","Unnamed: 0"],axis=1).set_index("調査ID")
pcdf = pcdf.T
pcdf = pcdf.fillna(0)
phonedf = phonedf.T
phonedf = phonedf.fillna(0)
```

```python
for id in pcids:
    if id not in phoneids:
        phonedf = pd.concat([pcdf[id],phonedf],axis=1)
    else:
        print(id)
        phonedf[id] = phonedf[id] + pcdf[id]
```
